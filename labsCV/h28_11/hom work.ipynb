```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = (18,12)
```

```python
img = cv2.imread('1.jpg',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
hist = cv2.calcHist([img],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist)
```

```python
img_equalized = cv2.equalizeHist(img)
plt.imshow(img_equalized,cmap='gray')
```

```python
hist = cv2.calcHist([img_equalized],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist)
```

```python
img1 = cv2.imread('4.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('5.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python

img1_shape=img1.shape
img2_shape=img2.shape
print(img1_shape,img2_shape)
```

```python
img1_img2 = cv2.addWeighted(src1=img1,alpha=0.7,src2=img2,beta=0.9,gamma=-50)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.imshow(img1)
ax2.imshow(img2)
ax3.imshow(img1_img2)
```

```python
img1 = cv2.imread('logo.jpg')
img1 =cv2.resize(img1,(100,100))
img2 = cv2.imread('background.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img3 = img2.copy()
img3[10 : 10 + img1.shape[0], 10 : 10 + img1.shape[1]] = img1
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.imshow(img1)
ax2.imshow(img2)
ax3.imshow(img3)
```

```python
img = cv2.imread('book.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
pts1 = np.float32([[380,50],[1100,50],[200,600],[1400,600]])
pts2 = np.float32([[0,0],[1000,0],[0,550],[1000,550]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(1000,550))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img)
ax2.imshow(dst)
```
