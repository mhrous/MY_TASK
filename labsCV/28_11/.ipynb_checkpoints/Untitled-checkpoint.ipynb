```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (18,12)
```

```python
img1 = cv2.imread("1.jpg")
img2 = cv2.pyrUp(img1)
img3= cv2.pyrDown(img1)



```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.imshow(img1)
ax2.imshow(img2)
ax3.imshow(img3)
```

```python

img_s1 =cv2.resize(img1,None,fx=3,fy=3,interpolation=cv2.INTER_NEAREST)

img_s2 =cv2.resize(img1,None,fx=3,fy=3,interpolation=cv2.INTER_LINEAR)
img_s3 =cv2.resize(img1,None,fx=3,fy=3,interpolation=cv2.INTER_CUBIC)
img_s4 =cv2.resize(img1,None,fx=3,fy=3,interpolation=cv2.INTER_LANCZOS4)
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
ax1.imshow(img_s1)
ax2.imshow(img_s2)
ax3.imshow(img_s3)
ax4.imshow(img_s4)
```
